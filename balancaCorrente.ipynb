```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
#dados brutos
grav=9.8
corrBcte  = np.array([0.517, 1.054, 1.514, 2.026, 2.521, 3.008, 3.201]) #A
forcaBcte = np.array([-1.56, -1.69, -2.75, -2.85, -3.84, -4.96, -7.14])*grav*10**(-3) #g * m*s^2 * kilo = N

cmagIcte  = np.array([ 536,    451,   370,   291,   221,   130,    45,     4])#mT
forcaIcte = np.array([-6.78, -7.29, -7.80, -8.12, -8.07, -7.77, -7.57, -7.39])*grav*10**(-3) #g * m*s^2 * kilo = N

#regressão linear de corrBcte com forcaBcte para descobrir o slope e achar L 
resBcte = stats.linregress(corrBcte[:-1], forcaBcte[:-1])

#regressão linear de cmagIcte com forcaIcte para descobrir o slope e achar L 
resIcte = stats.linregress(cmagIcte, forcaIcte)

```

```python
plt.plot(corrBcte, forcaBcte, 'ok')
plt.plot(corrBcte, resBcte.intercept + resBcte.slope*corrBcte, color='red')
plt.xlabel('Corrente (A)')
plt.ylabel('Força (N)')
print(resBcte.slope)
```

```python
plt.plot(cmagIcte, forcaIcte, 'ok')
plt.plot(cmagIcte, resIcte.intercept + resIcte.slope*cmagIcte, color='red')
plt.xlabel('Campo Magnético (mT)')
plt.ylabel('Força (N)')
print(resIcte.slope)

#regressão polinomial 
polres = np.polyfit(cmagIcte, forcaIcte, 2)
plt.plot(cmagIcte, polres[0]*cmagIcte**2+ polres[1]*cmagIcte+ polres[2], '--')
```
